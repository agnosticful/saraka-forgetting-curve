```python
import numpy as np
from matplotlib import pyplot

# judgement (good:True, bad:False), last_interval (h), last_intareval_rate (%)
# return 
#     interval(h)
#     interval_rate(%)
def get_interval_togashing(judgement, last_interval, last_interval_rate):
    if judgement == False:

        interval = 24
        interval_rate = last_interval_rate - 20
        
    else:

        interval = last_interval * (last_interval_rate / 100)
        interval_rate = last_interval_rate

    return interval, interval_rate

def check_interval(judgements):

    pyplot.title('Togashing Curve')
    pyplot.xlabel('times')
    pyplot.ylabel('hours / rates')
    
    interval = 15
    interval_rate = 250
    hours = []
    interval_rates = []
    
    hours.append(interval)
    interval_rates.append(interval_rate)
    
    for judgement in judgements:
        
        got_interval = get_interval_togashing(judgement, interval, interval_rate)
        
        hours.append(got_interval[0])
        interval_rates.append(got_interval[1])
        
        interval = got_interval[0]
        interval_rate = got_interval[1]

    times = [i for i in range(len(hours))]
    
    pyplot.plot(times, hours, label='hours')
    pyplot.plot(times, interval_rates, label='rates')
    pyplot.show()
    
    print("times:",times)
    print("hours:",hours)
    print("interval_rates:",interval_rates)
```

```python
check_interval([True, True,True,True,True,True])
```

```python
check_interval([True, False,True,True,True,True])
```

```python
check_interval([True, False,True,False,True,True])
```
